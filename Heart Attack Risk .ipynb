### Polynomial Features

```python
# polynomial featurelar için yeni bir dataset oluşturuyoruz
poly_features = train_df[['cp', 'thalachh', 'slp','restecg','output']]

from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

poly_target = poly_features['output']

poly_features = poly_features.drop(columns = ['output'])

poly_features = imputer.fit_transform(poly_features)
from sklearn.preprocessing import PolynomialFeatures
                                  
# polynomialfeature derecesi 3
poly_transformer = PolynomialFeatures(degree = 3)
```

```python
poly_features.size
```

```python
poly_transformer.fit(poly_features)
poly_features = poly_transformer.transform(poly_features)
print('Polynomial Features shape: ', poly_features.shape)
```

```python
poly_transformer.get_feature_names(input_features = ['cp', 'thalachh', 'slp','restecg'])[:18]
```

```python
poly_features = pd.DataFrame(poly_features, 
                             columns = poly_transformer.get_feature_names(['cp', 'thalachh', 'slp','restecg','output']))

poly_features['output'] = poly_target

# output'a göre korelasyonlar
poly_corrs = poly_features.corr()['output'].sort_values()

# En pozitif ve en negatif korelasyonlar
print(poly_corrs.head(10))
print(poly_corrs.tail(5))
```

```python
poly_features
```

### Concat two data (poly_features and train_df)

```python
new_data = train_df.drop(columns = ['cp', 'thalachh', 'slp','restecg'])
```

```python
app_train_poly =  pd.concat([poly_features, new_data], axis=1, join='inner')
```

```python
app_train_poly
```

```python
print('Training data shape: ', app_train_poly.shape)
```

```python
target_values
```

```python
app_train_poly = app_train_poly.drop(columns = ['output'])
```

### Train-test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(app_train_poly,target_values, test_size = 0.2, random_state = 42)
print("The shape of X_train is      ", X_train.shape)
print("The shape of X_test is       ",X_test.shape)
print("The shape of y_train is      ",y_train.shape)
print("The shape of y_test is       ",y_test.shape)
#train-test kısımlarını ayırıyoruz
```

### Logistic Regression 

```python
from sklearn.metrics import accuracy_score, classification_report, roc_curve
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
logreg = LogisticRegression()

# eğitim
logreg.fit(X_train, y_train)

# olasılık hesabı
y_pred_proba = logreg.predict_proba(X_test)

# tahmin edilen olasılığın sonucunu bulma
y_pred = np.argmax(y_pred_proba,axis=1)

# test accuracy
print("The test accuracy score of Logistric Regression is ", accuracy_score(y_test, y_pred))
```

### Gradient Boosting Classifier

```python

gbt = GradientBoostingClassifier(n_estimators = 50,max_depth=1,subsample=0.8,max_features=0.25,random_state=42)

gbt.fit(X_train,y_train)

y_pred = gbt.predict(X_test)
print("The test accuracy score of Gradient Boosting Classifier is ", accuracy_score(y_test, y_pred))
```

### SVC 

```python
clf = SVC(kernel='linear', C=1, random_state=42).fit(X_train,y_train)

y_pred = clf.predict(X_test)

print("The test accuracy score of SVM is ", accuracy_score(y_test, y_pred))
```

### Decision Tree Classifier 

```python
dt = DecisionTreeClassifier(random_state = 42)

dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

print("The test accuracy score of Decision Tree is ", accuracy_score(y_test, y_pred))
```

### Random Forest Classifier

```python

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred = dt.predict(X_test)

print("The test accuracy score of Random Forest is ", accuracy_score(y_test, y_pred))
```

### Cat Boost Classifier

```python
from catboost import CatBoostClassifier
cbc = CatBoostClassifier(
    iterations=10, 
    learning_rate=0.29, 
    #loss_function='CrossEntropy'
)

cbc.fit(X_train,y_train)
y_pred=cbc.predict(X_test)
print("CatBoostClassifier Accuracy : {} ".format(accuracy_score(y_test,y_pred)))
```
